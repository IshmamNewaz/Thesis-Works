#Correlations

```python
data = pd.read_csv('/content/GlobalTemperatures.csv')
data = data[['LandAverageTemperature','LandMaxTemperature','LandMinTemperature', 'LandMaxTemperatureUncertainty','LandMinTemperatureUncertainty']]
data
```

```python
correlation = data.corr()
correlation
```

```python
sns.heatmap(correlation)
```

```python
sns.heatmap(correlation, cmap='RdBu', vmax=1, vmin=0, annot=True, square = True,
            annot_kws={'fontsize':11, 'fontweight':'bold'})
```
