```python
import numpy as np
from statsmodels.stats.inter_rater import fleiss_kappa
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Example rating matrix (rows = items, columns = categories)
# Each cell represents the number of raters who rated the item in the respective category
rating_matrix = np.array([
    [0, 0, 10],  # All 10 raters assigned to category 3
    [2, 8, 0],   # 2 raters assigned to category 1, 8 to category 2
    [5, 5, 0],   # 5 raters assigned to category 1, 5 to category 2
    [0, 10, 0],  # All 10 raters assigned to category 2
    [3, 4, 3]    # Mixed ratings among categories
])

# Calculate Fleiss' Kappa
kappa = fleiss_kappa(rating_matrix, method='fleiss')

# Output the result
print("Fleiss' Kappa:", kappa)
```

```python
sns.heatmap(rating_matrix, annot=True, cmap="YlGnBu", cbar=False, fmt="d")
plt.xlabel("Categories")
plt.ylabel("Items")
plt.title("Rating Matrix")
plt.show()
```
