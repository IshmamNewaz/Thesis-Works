```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
path = '/content/aerial.jpg'  # Update with your image path
img = cv2.imread(path, 0)  # Read image in grayscale

# 1. Create a heatmap based on pixel intensity values
# You can use the image itself or derive some feature, like gradients or edge maps.

# Create a heatmap from the image itself (e.g., use pixel intensities as heatmap)
heatmap = img.astype(np.float32)

# Normalize the heatmap (between 0 and 1) for better visualization
heatmap = cv2.normalize(heatmap, None, 0, 1, cv2.NORM_MINMAX)

# 2. Create the heatmap overlay (apply colormap for better visual representation)
heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap), cv2.COLORMAP_JET)

# 3. Convert the grayscale image `img` to 3 channels for compatibility with `heatmap_color`
img_colored = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

# 4. Overlay the heatmap on the image
overlay = cv2.addWeighted(img_colored, 0.7, heatmap_color, 0.3, 0)

# 5. Display the results
# First, display the original image
plt.figure(figsize=(10, 6))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')  # Hide axis
plt.show()

# Then, display the heatmap overlay on top of the image
plt.figure(figsize=(10, 6))
plt.imshow(overlay)
plt.title("Heatmap Overlay")
plt.axis('off')  # Hide axis
plt.show()

# If you want to show the heatmap separately as well:
plt.figure(figsize=(10, 6))
plt.imshow(heatmap_color)
plt.title("Heatmap (Colorized)")
plt.axis('off')  # Hide axis
plt.show()
```
