```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/aerial.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
if image is None:
    print("Error: Image not found!")
    exit()
```

```python
_, binary_mask = cv2.threshold(image, 125, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

# Binary segmentation map
plt.subplot(1, 2, 2)
plt.title("Binary Segmentation Map")
plt.imshow(binary_mask, cmap='gray')
plt.axis('off')

plt.show()

# Save the binary mask image (optional)
cv2.imwrite('binary_segmentation_map.jpg', binary_mask)
```
